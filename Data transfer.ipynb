## Importing data to sql 

```python
import pandas as pd 
from sqlalchemy import create_engine
```

```python
df=pd.read_csv("Global_Superstore.csv")
df.head()
```

```python
df.columns=df.columns.str.lower().str.replace(" ","_").str.replace("-","_")
```

```python
df["order_date"]=pd.to_datetime(df["order_date"])
df["ship_date"]=pd.to_datetime(df["ship_date"])
df
```

```python
engine = create_engine('postgresql://postgres:postgres123@127.0.0.1:5433/Global_Superstore')

df.to_sql(
        'gs',
        engine,
        if_exists='replace',  
        index=False,
        chunksize=1000,  
        method='multi')
```
